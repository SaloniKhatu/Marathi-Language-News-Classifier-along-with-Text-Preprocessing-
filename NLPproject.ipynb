```python
import stanza
import pandas as pd
from cltk.stop.marathi.stops import STOP_LIST as stop_words_marathi
import re
stanza.download('mr')
nlp = stanza.Pipeline('mr', processors='tokenize,pos,lemma,ner,mwt', verbose=True, use_gpu=True)
```

```python
print(stop_words_marathi[:50])
```

```python
def preprocessing(input_text):
  
  input_text = re.sub('[-।,:?;!\(\)]','',input_text)
  input_text = " ".join(input_text.split())

  doc = nlp(input_text)
  """
  print("Word Segmentation and stopword removal")
  for i, sentence in enumerate(doc.sentences):
    print(f'====== Sentence {i+1} tokens =======')
    print(*[f'id: {token.id}\ttext: {token.text}' for token in sentence.tokens if token not in stop_words_marathi], sep='\n')
  
  print("\nPOS Tagging")
  print(*[f'word: {word.text}\tupos: {word.upos}\tfeats: {word.feats if word.feats else "_"}' for sent in doc.sentences for word in sent.words], sep='\n')
  #upos link: https://universaldependencies.org/u/pos/

  print("\nLemmatization")
  print(*[f'word: {word.text+" "}\tlemma: {word.lemma}' for sent in doc.sentences for word in sent.words], sep='\n')
  """
  #print(doc)
  for i, sentence in enumerate(doc.sentences):
    tokenized_text = [[token.id, token.text] for token in sentence.tokens]
  postagged_text = [[word.text,word.upos,word.feats] for sent in doc.sentences for word in sent.words]
  lemmatized_text = [[word.text,word.lemma] for sent in doc.sentences for word in sent.words]

  return tokenized_text, postagged_text, lemmatized_text
```

```python
def clean(text):
  text_lemma = []
  doc = nlp(text)
  lemma = []
  for sentence in doc.sentences:
      for word in sentence.words:
          text_lemma.append(word.lemma)
  text_lemma = " ".join(word for word in text_lemma)
  return text_lemma
```

```python
text = """यूएईने टीम इंडियाला संकटातून वाचवले, विश्वविजेत्या संघाला दुसऱ्या गटात पाठवले; पाहा सुधारित वेळापत्रक"""
tokenized_text, postagged_text, lemmatized_text = preprocessing(text)
print("\nTokenization")
for i in tokenized_text:
  print("ID: {0}\t Token: {1}".format(i[0],i[1]))
print("\nPOS Tagging")
for i in postagged_text:
  print("Word: {0}\t Universal POS: {1}\t Features: {2}".format(i[0],i[1],i[2]))
print("\nLemmatization")
for i in lemmatized_text:
  print("Word: {0}\t Lemma: {1}".format(i[0],i[1]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('news.csv')

X = df['headline']
y = df['label']
 
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=104, 
                                   test_size=0.25, 
                                   shuffle=True)
```

```python
X_train_cleaned = X_train.map(lambda x: clean(x))
X_test_cleaned = X_test.map(lambda x: clean(x))
```

```python
X_train_cleaned[0]
```

```python
vectorizer = TfidfVectorizer()
X_train_Vectorized = vectorizer.fit_transform(X_train_cleaned)
X_test_Vectorized = vectorizer.transform(X_test_cleaned)
```

```python
import pickle
pickle.dump(vectorizer, open("vectorizer.pickle", "wb"))
```

```python
classifier = MultinomialNB()
classifier.fit(X_train_Vectorized, y_train)
```

```python
Y_pred = classifier.predict(X_test_Vectorized)
```

```python
print('We got an accuracy of',np.mean(Y_pred == y_test)*100, '% over the test data.')
```

```python
cf_matrix = confusion_matrix(y_test,Y_pred)

print("Confusion Matrix \n",cf_matrix,"\n")
print("Accuracy is",round(accuracy_score(y_test,Y_pred),2)*100,"%")
print("F1 Score is",f1_score(y_test,Y_pred,average='weighted'))
```
